```python
def read_file(path, halo_num):
    '''Function to read in the timestep bulk-processing datafile (from /home/akinhol/Data/Timescales/DataFiles/{name}.data)'''
    data = []
    with open(path,'rb') as f:
        while True:
            try: 
                data.append(pickle.load(f))
            except EOFError:
                break
    data = pd.DataFrame(data)
    
    if not halo_num=='all':
        data = data[data.z0haloid == halo_num]

    return data
```

```python
from tangos_halo_module.halo_properties import get_main_progenitor_branch, track_halo_property, get_timesteps, ID_to_sim_halo_snap, infall_final_n_particles, infall_final_coordinates, apocentric_distance, disruption_time, accretion_time, orbit_interpolation, infall_velocity, max_sSFR_time, max_mass_time 
from tangos_halo_module.path import get_file_path, get_halo_snap_num, read_file
from tangos_halo_module.halos import ID_to_tangos_halo, get_survivors, get_zombies, get_host, get_survivor_IDs, get_zombie_IDs, blockPrint, enablePrint, tangos_to_pynbody_halo

def quenching_time(ID=0, simulation=0, status=0, tangos_halo=0, halo_id=0, snap_num=0, threshold=2e-11):
    '''
    input params: 
        ID: new sim/snap/halo id
        simulation: h148, h229, h242, h329: string
        status: 'Zombie', 'Survivor': string
        tangos_halo: 0 or a valid Tangos halo object
        halo_id: halo id: 0 or string or numeric
        snap_num: 0 or 4 digit simulation snapshot number: string
        threshold: threshold in sSFR below which a galaxy is quenched
            [units: /yr]
    output params:
        quenching time: earliest time a satellite falls below the threshold sSFR 
            [units: Gyr since start of simulation]
    '''
    import pandas as pd
    import h5py
    d = pd.read_csv('sSFR_data.csv')
    ids = d['IDs'].to_numpy()
    sSFRs = d['sSFR'].to_numpy()
    Mstar = d['stellar_mass'].to_numpy()
#     Mstar = d['Mass'].to_numpy()
#     new_Mstar = d['']
    
    #np.seterr(divide='ignore')
    if ID != 0:
        simulation, status, halo_id, snap_num = ID_to_sim_halo_snap(ID=ID)
        
    #Host
    path = get_file_path(tangos_halo=0, simulation=simulation, status='Host', halo_id='1', snap_num='4096')
    with h5py.File(path, 'r') as fh:
        t_Gyr = fh['time_|_Gyr'][:]
    sSFR = sSFRs[ids==ID][0] # A string of form [a b c ...]
    Mstar = Mstar[ids==ID][0]
    #print(sSFR)
    sSFR = np.array(list(map(float, sSFR[1:-1].split())))/1e6
    Mstar = np.array(list(map(float, Mstar[1:-1].split())))
    #print(sSFR)
    # Replace nan and inf values with 0
    # When making log plots of sSFR
    
    t_infall = accretion_time(ID=ID, simulation=simulation, status=status, tangos_halo=tangos_halo, 
                              halo_id=halo_id, snap_num=snap_num, category='stars')[1]
    if status=='Survivor':
        t_final_sSFR = t_Gyr[-1]
    else:
        t_final_sSFR = max(t_Gyr[sSFR>=0])
    final_sSFR = sSFR[t_Gyr==t_final_sSFR]
    if final_sSFR<=threshold and final_sSFR>=0: 
        #quenched at z=0 or final snap before disruption
            #z=0 or final stellar mass is non-zero
        #find times when the sSFR transitioned from above to below the threshold
        dips = np.zeros(len(sSFR))
        for i, val in enumerate(sSFR):
            if i>0 and sSFR[i]<=threshold and sSFR[i-1]>threshold and sSFR[i]>=0:
                #sSFR at current timestep is below threshold
                    #sSFR at previous timestep was above threshold
                        #the satellite has not merged with host
                            #current stellar mass is nonzero
                                #current sSFR is non-negative
                dips[i]=1
        if len(t_Gyr[dips==1]) != 0:
            #the galaxy has transitioned from star-forming to quiescent at least once before z=0
            #recall that these galaxies are quenched at z=0. so these galaxies have always been below the threshold
            #find the latest time that a galaxy crossed from star-forming to quiescent
            quenching_time = max(t_Gyr[dips==1]) 
        else:
            #this survivor has never transitioned from star-forming to quiescent before z=0
            #recall that these galaxies are quenched at z=0
            #find earliest time the galaxy met the sSFR threshold
            # Check how many survivors fall here: many do :((
            quenching_time = min(t_Gyr[(sSFR<=threshold) & (sSFR>=0)])
    else: 
        #this survivor is star-forming at z=0
        quenching_time = -1

#     else:
#         #quenching_time = -1
#         raise valueError('Halo must be a Survivor or Zombie. Try status = "Zombie" or "Survivor".')
    
    return quenching_time, sSFR, Mstar
```

```python
unmatched = pd.DataFrame()
path = 'ChristensenData/'
d = pd.read_csv('Data.csv')
k = 5
myID = []
hollisID = []
sims = []
my_tquench = []
my_tquench_lbk = []
hollis_tquench = []
my_nstar = []
hollis_nstar = []
for sim in ['h148', 'h229', 'h242', 'h329']:
    print(sim)
    my_ids = d.IDs[(d.Simulation==sim)&(d.Status=='Survivor')]
    # Hollis' data:
    data = read_file(path+sim+'.data','all')
    hostx = data[data.haloid==1].Xc[0]
    hosty = data[data.haloid==1].Yc[0]
    hostz = data[data.haloid==1].Zc[0]
    for ID in my_ids:
        myID.append(ID)
        sims.append(sim)
        # My quenching time
        #tq = d[(d.IDs==ID)].iloc[0]['time_quench']
        tq, sSFR, Mstar=quenching_time(ID=ID)
        my_tquench.append(tq)
        if tq >= 0:
            my_tquench_lbk.append(13.800797497330507 - tq)
        else:
            my_tquench_lbk.append(-1)
        my_nstar.append(d[(d.IDs==ID)].iloc[0]['final_n_star'])
        print('My ID: ' + str(ID))
        myX = d[(d.Simulation==sim)&(d.Status=='Survivor')&(d.IDs==ID)].iloc[0]['final_x']
        myY = d[(d.Simulation==sim)&(d.Status=='Survivor')&(d.IDs==ID)].iloc[0]['final_y']
        myZ = d[(d.Simulation==sim)&(d.Status=='Survivor')&(d.IDs==ID)].iloc[0]['final_z']
        space_cut = (data.Xc-hostx <= myX+k) & (data.Xc-hostx >= myX-k) & (data.Yc-hosty <= myY+k) & (data.Yc-hosty >= myY-k) & (data.Zc-hostz <= myZ+k) & (data.Zc-hostz >= myZ-k)
        HX = data[space_cut & (data.haloid!=1)][['haloid']]
        if len(HX)==0:
            print('Hollis ID: No Match')
            hollisID.append(-1)
            hollis_tquench.append(-1)
            hollis_nstar.append(-1)
            unmatched = pd.concat([unmatched, d[(d.IDs==ID)]])
        else:
            HX = data[space_cut & (data.haloid!=1)][['haloid']].iloc[0]['haloid']
            hollisID.append(HX)
            HQ = np.asarray([data[data.haloid==HX].iloc[0]['tquench']])[0]
            hollis_tquench.append(HQ)
            HS = np.asarray([data[data.haloid==HX].iloc[0]['n_star']])[0]
            hollis_nstar.append(HS)
            print('Hollis ID: ' + str(HX))
            print('My Q: ', tq, ' Hollis Q: ', HQ)
            print(sSFR)
            print(Mstar)
print(myID)
print(hollisID)
print(sims)
print(hollis_tquench)
print(my_tquench)
print(my_tquench_lbk)
print(my_nstar)
print(hollis_nstar)
unmatched
```
