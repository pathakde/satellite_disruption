```python
# stack separately zombies and survivors with and without (dummy) array
plt.rcParams.update({'figure.max_open_warning': 0})

h = 0.6776942783267969
sim='h329'
host = get_host(simulation=sim)
a = get_timesteps(simulation=sim)[2]
t_Gyr = get_timesteps(simulation=sim)[0]
MPB = get_main_progenitor_branch(tangos_halo=host, simulation=sim)[0]
for i, MW in enumerate(MPB):
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharex='none', sharey='none', figsize=(24, 8))
    fig.set_facecolor('w')
    np.seterr(divide='ignore')
    plt.setp(ax1.spines.values(), linewidth=3)
    plt.setp(ax2.spines.values(), linewidth=3)
    plt.setp(ax3.spines.values(), linewidth=3)
    
    hx=MW['Xc'][0]*a[i]/h
    hy=MW['Yc'][0]*a[i]/h
    hz=MW['Zc'][0]*a[i]/h
    if 'childHalo' in MW.keys():
        satellites=MW['childHalo']
        for sat in satellites:
            sx=sat['Xc'][0]*a[i]/h
            sy=sat['Yc'][0]*a[i]/h
            sz=sat['Zc'][0]*a[i]/h
            X=sx-hx
            Y=sy-hy
            Z=sz-hz
            ax1.scatter(X, Y, s=400, alpha=1, marker='.', edgecolor='k', facecolor='orangered', lw=1)
            ax2.scatter(Y, Z, s=400, alpha=1, marker='.', edgecolor='k', facecolor='orangered', lw=1)
            ax3.scatter(Z, X, s=400, alpha=1, marker='.', edgecolor='k', facecolor='orangered', lw=1)
    #fig.suptitle('t = '+str(t_Gyr[i])+' Gyr', fontsize=23)
     
    ax1.scatter(0, 0, s=500, alpha=1, marker='*', edgecolor='k', lw=2)
    ax1.set_title('X-Y', fontsize=23)
    ax1.set_xlabel('X [kpc]', fontsize=23)
    ax1.set_ylabel('Y [kpc]', fontsize=23)
    ax1.set_xlim(-200, 200)
    ax1.set_ylim(-200, 200)
    ax1.tick_params(which='major', length=12)
    ax1.tick_params(direction='in', which='both', labelsize=20, bottom=True, top=True, left=True, right=True)
    
    ax2.scatter(0, 0, s=500, alpha=1, marker='*', edgecolor='k', lw=2, label='t = '+str("%.3f" % t_Gyr[i])+' Gyr')
    ax2.set_title('Y-Z', fontsize=23)
    ax2.set_xlabel('Y [kpc]', fontsize=23)
    ax2.set_ylabel('Z [kpc]', fontsize=23)
    ax2.set_xlim(-200, 200)
    ax2.set_ylim(-200, 200)
    ax2.tick_params(which='major', length=12)
    ax2.tick_params(direction='in', which='both', labelsize=20, bottom=True, top=True, left=True, right=True)
    ax2.legend(loc='upper center', fontsize=23, frameon=True)
    
    ax3.scatter(0, 0, s=500, alpha=1, marker='*', edgecolor='k', lw=2)
    ax3.set_title('Z-X', fontsize=23)
    ax3.set_xlabel('Z [kpc]', fontsize=23)
    ax3.set_ylabel('X [kpc]', fontsize=23)
    ax3.set_xlim(-200, 200)
    ax3.set_ylim(-200, 200)
    ax3.tick_params(which='major', length=12)
    ax3.tick_params(direction='in', which='both', labelsize=20, bottom=True, top=True, left=True, right=True)
    
    fig.tight_layout()
    if len(str(i))==1:
        filename='position_img/h329/rad_distr0'+str(i)+'.png'
    else:
        filename='position_img/h329/rad_distr'+str(i)+'.png'
    fig.savefig(filename, dpi=None, facecolor='w', edgecolor='w',
                orientation='portrait', format='png',
                transparent=False, bbox_inches=None, pad_inches=0.1,
                metadata=None)  
    plt.show()
```
